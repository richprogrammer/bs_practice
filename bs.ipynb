```python
import requests
url = "https://olympics.com/ioc/national-olympic-committees"
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:99.0) Gecko/20100101 Firefox/99.0"}
r = requests.get(url=url, headers=headers)  
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.content)
```

```python
soup
```

```python
print(soup.prettify())
```

```python
div = soup.find('div', attrs={"class":"sc-iIgkCN edLvuQ"})
```

```python
countries = div.findAll('div', attrs={'data-cy':'noc-item'})
countries
```

```python
len(countries)
```

```python
countries[0]
```

```python
countries[0]['href']
```

```python
countries[0].img
```

```python
# Assuming countries is a list of BeautifulSoup tags with 'href' attributes

# Initialize an empty list to store the modified hrefs
modified_hrefs = []

# Loop through each country tag in the list
for country in countries:
    # Replace '/ioc/' with an empty string in the href attribute and append to the list
    modified_hrefs.append(country['href'].replace('/ioc/', ''))

# Now, modified_hrefs contains all the hrefs with '/ioc/' removed
```
